## 생존자 비율

* 전체 승객 중 남성 생존자 비율, 여성 생존자 비율
    * 남성 생존자 비율 = (남성 생존자 수) / (전체 승객 수)
    * 여성 생존자 비율 = (여성 생존자 수) / (전체 승객 수)

* 생존자들 중 남성 생존자 비율, 여성 생존자 비율
    * 남성 생존자 비율 = (남성 생존자 수) / (전체 생존자 수)
    * 여성 생존자 비율 = (여성 생존자 수) / (전체 생존자 수)

* 생존자들 중 성인 생존자 비율, 어린이 생존자 비율
    * 성인 생존자 비율 = (성인 생존자 수) / (전체 생존자 수)
    * 어린이 생존자 비율 = (어린이 생존자 수) / (전체 생존자 수)

```python
msurvived_titanic = titanic[(titanic.sex == 'male')&(titanic.survived == 1)]
fsurvived_titanic = titanic[(titanic.sex == 'female')&(titanic.survived == 1)]

malerate = len(msurvived_titanic) / len(titanic)
femalerate = len(fsurvived_titanic) / len(titanic)

print(f'남성 생존자 비율: {malerate * 100}')
print(f'여성 생존자 비율: {femalerate * 100}')
```

```python
survived = titanic[titanic.survived == 1]

male_survived = len(msurvived_titanic) / len(survived)
female_survived = len(fsurvived_titanic) / len(survived)

print(f'남성 생존자 비율: {male_survived * 100}')
print(f'여성 생존자 비율: {female_survived * 100}')
```

```python
adult_suv = titanic[(titanic.age >= 18)&(titanic.survived == 1)]
child_suv = titanic[(titanic.age < 18)&(titanic.survived == 1)]

adult_suvrate = len(adult_suv) / len(survived)
child_suvrate = len(child_suv) / len(survived)

print(f'성인 생존자 비율: {adult_suvrate * 100}')
print(f'어린이 생존자 비율: {child_suvrate * 100}')
```

## 객실 등급별 생존자 비율

* (1등급 객실 생존자 비율) = (1등급 객실 생존자 수) / (전체 생존자 수)
* (2등급 객실 생존자 비율) = (2등급 객실 생존자 수) / (전체 생존자 수)
* (3등급 객실 생존자 비율) = (3등급 객실 생존자 수) / (전체 생존자 수)

```python
first_suv = titanic[(titanic.pclass == 1)&(titanic.survived == 1)]
second_suv = titanic[(titanic.pclass == 2)&(titanic.survived == 1)]
third_suv = titanic[(titanic.pclass == 3)&(titanic.survived == 1)]

first_suvrate = len(first_suv) / len(survived)
second_suvrate = len(second_suv) / len(survived)
third_suvrate = len(third_suv) / len(survived)

print(f'1등급 객실 생존자 비율: {int(first_suvrate * 100)}%')
print(f'2등급 객실 생존자 비율: {int(second_suvrate * 100)}%')
print(f'3등급 객실 생존자 비율: {int(third_suvrate * 100)}%')
```

* (1등급 객실 생존자 비율) = (1등급 객실 생존자 수) / (1등급 객실 승객수)
* (2등급 객실 생존자 비율) = (2등급 객실 생존자 수) / (2등급 객실 승객수)
* (3등급 객실 생존자 비율) = (3등급 객실 생존자 수) / (3등급 객실 승객수)

```python
first = titanic[titanic.pclass == 1]
second = titanic[titanic.pclass == 2]
third = titanic[titanic.pclass == 3]

first_suvrate = len(first_suv) / len(first)
second_suvrate = len(second_suv) / len(second)
third_suvrate = len(third_suv) / len(third)

print(f'1등급 객실 생존자 비율: {first_suvrate * 100}')
print(f'2등급 객실 생존자 비율: {second_suvrate * 100}')
print(f'3등급 객실 생존자 비율: {third_suvrate * 100}')
```

```python
for c in titanic.pclass.unique():
    # 각 객실 등급의 승객 수
    num_of_class = len(titanic[titanic.pclass == c])
    # 각 객실 등급의 생존자 수
    num_of_class_alive=len(titanic[(titanic.pclass == c)&(titanic.alive == 'yes')])
    # 각 객실 등급에서 생존자 비율
    print(f'{c}등급 객실에서 생존자 비율: {num_of_class_alive/num_of_class*100}')
```
