## indexing, slicing
`iloc` vs `loc`
* `iloc`: 정수 인덱스 기반으로 원소를 참조. list의 인덱스를 사용하는 방법과 동일.
* `loc`: 축 레이블 기반으로 원소를 참조.

```python
s.iloc[0] # Series의 첫번째 원소 참조(선택)
```

```python
s.iloc[-1] # Series의 마지막 원소 참조(선택)
```

```python
s.iloc[:2] # slicing의 시작 인덱스 0은 생략 가능. 첫번째 원소부터 2번째 원소까지 slicing.
```

```python
s.iloc[-2:] # 마지막 2개 원소를 slicing. 음수 인덱스 사용 가능. 마지막 인덱스는 생략 가능.
```

```python
s.loc[3] # Series에서 레이블이 3인 원소를 참조.
```

```python
s.loc[-1]
```

```python
s.loc[:2]
```

## DataFrame에서 column 선택

* `df['column_name']`
* `df.column_name`
    * 컬럼 이름이 변수 이름 규칙에 맞지 않는 경우에는 사용할 수 없음.
        * 변수 이름은 영문자, 숫자, 밑줄(underscore, _)만 사용할 수 있음.
        * 변수 이름은 숫자로 시작할 수 없음.
        * 파이썬의 키워드(if, for, def, class, ...)들은 변수 이름으로 사용할 수 없음.
    * 컬럼 이름이 DataFrame이 가지고 있는 속성(변수, 메서드) 이름과 같은 경우에는 사용할 수 없음.

## boolean indexing

`df[조건식]`과 `df.loc[조건식]`은 같은 코드.
* 조건식을 만족하는 모든 행들을 선택.

```python
# 서울 데이터만 선택
df[df['city'] == '서울']
```

```python
# 2024년 데이터를 선택

df[df['year'] == 2024]
```

Series, DataFrame에서 boolean indexing을 사용할 때 주의할 점:
* and, or, not을 사용할 수 없음.
* 대신에 &, |, ~ 연산자를 사용.
* 반드시 ()를 사용해서 논리 연산의 순서를 명시해야 함.

```python
df
```

```python
# 2021, 2022년 데이터 선택

# df[(df['year'] == 2022) | (df['year'] == 2023)]

df[(df.year == 2022) | (df.year == 2023)]
```

```python
# 도시가 서울이거나 pop의 값이 1.2 이상인 데이터 선택

df[(df['city'] == '서울') | (df['pop'] >= 1.2)]
```
